**UNIVARIATE ANALYSIS (Single Variable)**
1. Histogram – Distribution of Layoffs

```python
plt.hist(df['total_laid_off'].dropna())
plt.title("Distribution of Total Layoffs")
plt.xlabel("Number of Employees Laid Off")
plt.ylabel("Frequency")
plt.show()
```

**Interpretation:**

Most companies have laid off a smaller number of employees, while a few companies have very high layoffs. This indicates that layoffs are concentrated among a few large companies.

2.Countplot – Industry Distribution

```python
sns.countplot(y=df['industry'], order=df['industry'].value_counts().index)
plt.title("Number of Layoff Records by Industry")
plt.show()
```

**Interpretation:**

The technology and finance industries have the highest number of layoff records, indicating these sectors are more frequently affected by workforce reductions.

3.Boxplot – Layoffs Spread

```python
sns.boxplot(x=df['total_laid_off'])
plt.title("Boxplot of Total Layoffs")
plt.show()
```

**Interpretation:**

The presence of outliers shows that some companies have extremely high layoffs compared to others. Most companies have relatively lower layoffs.

**BIVARIATE ANALYSIS (Two Variables)**

4. Barplot – Layoffs by Industry

```python
industry_layoffs = df.groupby('industry')['total_laid_off'].sum().sort_values(ascending=False)

industry_layoffs.head(10).plot(kind='bar')
plt.title("Top Industries by Total Layoffs")
plt.xlabel("Industry")
plt.ylabel("Total Layoffs")
plt.xticks(rotation=45)
plt.show()
```

**Interpretation:**

The technology industry has the highest layoffs, suggesting it is the most impacted sector. Other industries show comparatively lower workforce reductions.

5. Scatter Plot – Funding vs Layoffs

```python
plt.scatter(df['funds_raised'], df['total_laid_off'])
plt.title("Funding vs Layoffs")
plt.xlabel("Funds Raised")
plt.ylabel("Total Laid Off")
plt.show()
```

**Interpretation:**

There is no strong direct relationship between funding and layoffs, indicating that even well-funded companies may reduce workforce due to strategic decisions.

6. Correlation Heatmap

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

**Interpretation:**

The correlation between variables is generally low, suggesting layoffs are influenced by multiple independent factors rather than a single variable.

MULTIVARIATE ANALYSIS (3+ Variables)

7.Pairplot

```python
sns.pairplot(df[['total_laid_off','funds_raised']].dropna())
plt.show()
```

**Interpretation:**

The pairwise relationships show no strong linear trends, reinforcing that layoffs depend on multiple complex factors.

8. Pivot Table (Industry vs Country)

```python
pivot = pd.pivot_table(df, values='total_laid_off', index='industry', columns='country', aggfunc='sum')

print(pivot)
```

**Interpretation:**

Certain industries dominate layoffs in specific countries, indicating regional differences in economic conditions and industry impact.

9. Grouped Analysis (Year + Industry)

```python
df['year'] = df['date'].dt.year

grouped = df.groupby(['year','industry'])['total_laid_off'].sum().unstack()

grouped.plot(kind='line')
plt.title("Yearly Layoffs by Industry")
plt.show()
```

**Interpretation:**

Layoff trends vary by industry over time, with some industries experiencing spikes during specific years.
